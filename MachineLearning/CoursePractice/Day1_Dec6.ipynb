#### Types of ML:
1. **Supervised Learning**: Training ml model on sample labelled data. Post training, predictions are made from real-world scenarios or new/unseen data. Training based on input and output.

    a. **Regression**: Random number prediction(deals with continuous number 0 to infinity) - gold price prediction, stock price prediction, cricket score prediction, weather prediction

    b. **Classification**: categorical prediction - categorize where the data would fall into - india win or not, match cancelled or not, dog breed prediction, stock price will increase or not

2. **Unsupervised Learning**: No labelled data. training model based on data patterns. No output. Based on business usecase, data is categorized into a group.

    a. **Clustering**:from available data, identify hidden patterns and group them. When you see new data - categorize them into one of the groups based on the pattern. eg.: amazon products recommendations vary based on age, gender

   b. **Dimensionality Reduction**: Reduce extremely large data into compact data with minimum loss of information.

3. Semi-supervised Learning
4. **Reinforcement Learning**: more advanced than deep learning
    bots, agent > env > action > rewards
