# **File Operation - Read and write files**

Python privides built-in functions and methods to read from and write to files, both text and binary.

## **Text Files**

#### **1. Read a file**

```python
with open('files/exampleTxt.txt','r') as file:
    content = file.read()
    print(content)
```

read line by line

```python
with open('files/exampleTxt.txt','r') as file: #file opened in read mode 'r'
    for line in file:
        print(line)
```

```python
with open('files/exampleTxt.txt','r') as file:
    for line in file:
        print(line.strip()) #removes the new line character
```

#### **2. Write to a file**

```python
print("before file.write:")
with open('files/exampleTxt.txt','r') as file:
    for line in file:
        print(line.strip())

#over-writes the existing content of file with current content
with open('files/exampleTxt.txt','w') as file: #file opened in write mode 'w'
    file.write("Hello World\n")
    file.write("This is the second line\n")

print("\nafter file.write:")
with open('files/exampleTxt.txt','r') as file:
    for line in file:
        print(line.strip())
```

#### **3. Append data to a file**

```python
print("before file.write:")
with open('files/exampleTxt.txt','r') as file:
    for line in file:
        print(line.strip())

with open('files/exampleTxt.txt','a') as file: #file opened in append mode 'a'
    file.write("Append operation\nThis is the forth line\n")
    file.write("Hello World\n")

print("\nafter file.write:")
with open('files/exampleTxt.txt','r') as file:
    for line in file:
        print(line.strip())
```

#### **4. Copy data from one file and write in another**

```python
with open('files/exampleTxt.txt','r') as file:
    content = file.read()

with open('files/destinationTxt.txt','w') as file:
    file.write(content)
```

## **Binary files**

#### **1. Write to a file**

```python
data = b'\x00\x01\x02\x03\x04'
with open('files/exampleBinary.bin','wb') as file:
    file.write(data)
```

#### **2. Read data from file**

```python
with open('files/exampleBinary.bin','rb') as file:
    content = file.read()
    print(content)
```
