# **PYTHON BASICS**

##### **Syntax**:
refers to the set of rules that defines
the combination of symbols that are considered to be correctly structured program in a language.

##### **Semantics**:
refers to the meaning or interpretation of the symbol character and commands in a language. It is about what the code is supposed to do and it runs.

Syntax is how we are able to arrange the words so that we write a code, and semantics refers to the interpretation.

**3. Case Sensitivity**
                  Variable 'p' != Variable 'P'

```python
name = "name - Hello World"
Name = "Name - Hello World"
print(name)
print(Name)
```

**4. Indentation:**
            Indentation in Python is used to define the structure and hierarchy of the code. Unlike many other programming languages that uses braces to delimit block of codes, right. Python uses indentation to determine the grouping of statement.

```python
age = 32
if age > 18:
    print("Print triggered for within if loop - age > 18")
if age < 16:
    print("Print triggered for within if loop - age < 16")
```

**5. Line Continuation**
In case of line continuation, you can basically use a backslash, and continue the statement to the next line.

```python
total = 1+2+3\
    +4+5
print(total)
```


##### 6. Multiple Statements in one line

Separate each line with semi-colon ;

```python
x=5;y=9;z=x+y;print(z)
```
