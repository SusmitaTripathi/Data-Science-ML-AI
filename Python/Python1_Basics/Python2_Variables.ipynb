# Variables

Variables are fundamental elements in programming used to store data that can be referenced and manipulated in a program.
In Python, variables are created when you assign a value to them, and they do not need explicit declarationto reserve memory space.
The declaration happens automatically when you assign a value to a variable.

#### **Declaring and assigning values to variables**

```python
age = 32
height = 6.1
name = "Susmita Tripathi"
isStudent = False

#print

print("age:", age)
print("height:", height)
print("name:", name)
```

#### **Naming Convention:**

1. Variable name should be descriptive
2. Should always start with alphabet or underscore
3. Variable names can contain alphabets, numbers and underscore
4. They are case-sensitive

##### **Valid variable names:**
1. first_name
2. LastName
3. last1234
4. _last1234

##### **Invalid variable names:**
1. 2first_name
2. Last-Name
3. @last1234

#### **Understanding variable types**

Python is variably typed - type of variable is determined at runtime.

```python
age = 24 #int
height = 6.1 #float
name = "Susmita Tripathi" #str
isStudent = False #boolean

print(type(age))
print(type(height))
print(type(name))
print(type(isStudent))
```

#### **Type conversion**

```python
print(type(age))
age_str = str(age)
print(age_str)
print(type(age_str))
```

```python
age_string = '25'
age_int = int(age_string)
print(age_string)
```

```python
string = "String"
stringToInt = int(string)
print(stringToInt)
```

```python
print("height", height)
print(type(height))
print(int(height))
print(float(int(height)))
```

#### **Dynamic Typing**

Python allows the type of variable to change as the program executes

```python
var = 10
print(type(var))

var = 'Hello World'
print(type(var))

var = 3.14
print(type(var))
```
