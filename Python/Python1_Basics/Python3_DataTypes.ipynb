# **Data Types**

1. Data Types are a classification of data which tells the compiler or interpreter how the programmer intends to use the data.
2. They determine the type of operations that can be performed on the data, the values that teh data can take, and the amount of memory needed to store the data.

##### **Importance of data types in programming**

1. Data types ensure that data is stored in an efficient way.
2. They help in performing correct operations on data
3. Proper use of data types can prevent errors and bugs in the program.


##### **Basic data types**
1. Integers
2. Floating-point numbers
3. Strings
4. Boolean

##### **Advanced Data types**
1. List
2. Tuples
3. Sets
4. Dictionary




#### **1. Integers**

```python
age = 35
type(age)
```

#### **2. Floating-point numbers**

```python
width = 2.16
type(width)
```

#### **3. String**

```python
name = "Krish"
Name = 'Hello'
print(type(name))
print(type(Name))
```

#### **4. Boolean**

```python
isTrue = True
isFalse = bool() #assigns boolean value False to variable
print(type(isTrue))
print(isFalse)
print(type(isFalse))
```
