## **Arithmetic Operators**

```python
a = 10
b = 4

addResult = a+b
print("Addition: ",addResult)
subResult = a-b
print("Subtraction: ",subResult)
mulResult = a*b
print("Multiplication: ",mulResult)
divResult = a/b
print("Division: ", divResult)
floorDivResult = a//b
print("Floor Division Result: ",floorDivResult)
modResult = a%b
print("Modulus: ",modResult)
expResult = a**b
print("Exponent: ",expResult)
```

## **Logical Operators**

```python
x = True
y = False
```

#### **1. AND**

```python
print(x & y)
print(x & x)
print(y & y)
```

#### **2. OR**

```python
print(x | y)
print(x | x)
print(y | y)
```

#### **3. NOT**

```python
print(not x)
print(not y)
```

```python
#Simple calculator

firstNum = int(input("Enter first number: "))
secondNum = int(input("Enter second number: "))
sum = firstNum + secondNum
difference = firstNum - secondNum
product = firstNum * secondNum
division = firstNum / secondNum

print("First number:", firstNum)
print("Second number:", secondNum)
print("Sum:", sum)
print("Difference:", difference)
print("Product:", product)
print("Division:", division)
```
