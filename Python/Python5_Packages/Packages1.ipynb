# **Importing modules & packages in Python**

In python,modules and packages help organize and reuse code. Here's a comprehensive guide on how to import them.

#### **Syntax**:

import package_name

import package_name as alias_name

from package_name import function1,function2

from package_name import *

```python
import math #import math library

from Python.Python5_Packages.package.mathematics import subtraction, multiplication, division

math.sqrt(16)
```

importing particular methods

```python
from math import sqrt,pi
print(sqrt(16))
print(pi)
```

```python
import numpy as np
print(np.sqrt(16))
print(np.sqrt(pi))
```

```python
import csv

with open('example.csv',mode='w',newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["name","age","state"])
    writer.writerow(["Hello","19","World"])

with open('example.csv',mode='r',newline='') as file:
    reader = csv.reader(file)
    for row in reader:
        print(row)
```
