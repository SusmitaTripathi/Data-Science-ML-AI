# **Exception Handling**

Exception handling in python allows you to handle errors gracefully and take corrective measures without stopping the execution of the program.

#### **What are exceptions?**

Exceptions are events that disrupt the normal flow of the program. They occur when an error is encountered during program execution. Common exceptions include:
1. ZeroDivisionError: division by zero
2. FileNotFoundError: file not found
3. ValueError: Invalid error
4. TypeError: Invalid type

#### **Try except block**

Exception class is the parent class of all exceptions

```python
#custom exception handling
try:
    a=b
except ZeroDivisionError as e:
    print(e)
except Exception as ex:
    print(ex)
```

#### **Try-except-else**

Else block will be executed if no exception is caught in try block.

```python
try:
    a=b
except ZeroDivisionError as e:
    print(e)
except Exception as ex:
    print(ex)
else:
    print("a value:",a)
```

```python
try:
    a=20
except ZeroDivisionError as e:
    print(e)
except Exception as ex:
    print(ex)
else:
    print("a value:",a)
```

#### **Try-except-else-finally**

Execution of finally block does not depend on any exceptions in try block

```python
try:
    a=b
except ZeroDivisionError as e:
    print(e)
except Exception as ex:
    print(ex)
else:
    print("a value:",a)
finally:
    print("a value:",a)
```

```python
try:
    a=30
except ZeroDivisionError as e:
    print(e)
except Exception as ex:
    print(ex)
else:
    print("a value:",a)
finally:
    print("a value:",a)
```
