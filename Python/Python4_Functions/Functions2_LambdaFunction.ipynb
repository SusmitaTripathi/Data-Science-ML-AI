# Lambda Functions

Lambda functions are small anonymous functions defined using lambda keyword. They can have any number of arguments but only one expression. They are commonly used for short operations or as arguments to higher order functions.

#### Syntax

```python
lambda arguments: expression
```

```python
def addition(a,b):
    return a+b
```

```python
addition(1,2)
```

Above function only has one expression so can be replaced with lambda function.
lambda function is assigned to add variable. whenever required, add can be called to get the result

```python
add = lambda a,b: a+b
print(type(add))
```

calling lambda function

```python
add(2,3)
```

```python
#lambda function to check even or odd

evenOrOdd = lambda a: a%2==0

print(evenOrOdd(1))
print(evenOrOdd(2))
print(evenOrOdd(3))
```

```python
#lambda function for addition of 3 numbers

addThreeNo = lambda a,b,c: a+b+c

print(addThreeNo(1,2,3))
```
