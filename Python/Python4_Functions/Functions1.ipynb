Whether number is even or odd

```python
number = 24
if number%2 ==0:
    print("Number is even")
else:
    print("Number is odd")
```

```python
def even_or_odd(num):
    """Function to return even or odd number"""
    if num%2 ==0:
        print("Number is even")
    else:
        print("Number is odd")
```

##### **Function call**

```python
even_or_odd(5)
```

```python
even_or_odd(78)
```

#### **Variable Length Arguments**

**_Positional arguments_**

1. denoted using _***args**_
2. Comma separated argument list

```python
def print_numbers(*parameters): #positional arguments
    for parameter in parameters:
        print(parameter)
```

```python
print_numbers(1,2,9,"Krish")
```

**_Keywords Argument_**

1. denoted using _****kwargs**_
2. all parameters are defined in key-value pairs

```python
def print_details(**kwargs):
    for key, value in kwargs.items():
        print(key,": " ,value)
```

```python
print_details(name="hello", age=23, country="India")
```

```python
def print_args_kwargs(*args,**kwargs):
    print("Positional arguments:")
    for val in args:
        print(val)
    print("Keyword arguments:")
    for key, value in kwargs.items():
        print(key,": " ,value)
```

```python
print_args_kwargs(1,2,name="Susmita", age=23, country="India")
```

```python
print_args_kwargs(1,2,name="Susmita", age=23, country="India",4,"Hello")
```

#### Temperature Conversion:

```python
def tempConversion(temp,unit):
    if unit == 'C':
        return temp * 9/5 + 32
    elif unit == 'F':
        return (temp - 32) * 5/9
    else:
        return None

print(tempConversion(32,"F"))
print(tempConversion(0,'C'))
```

#### Password Strength Checker:

```python
def checkPasswordStrength(password):
    """Checks password strength:\n
    Takes input password and validates against pre-set conditions to check if password is strong.\n
    Input: password\n"""

    if len(password) < 8:
        return False
    elif len(password) > 16:
        return False
    if not any(char.isdigit() for char in password):
        return False
    if not any(char.islower() for char in password):
        return False
    if not any(char.isupper() for char in password):
        return False
    if not any(char in '!@#$%^&*_' for char in password):
        return False
    if any(char in '(){}[]|\\+=-;:\'\",<>.?/' for char in password):
        return False
    return True

print(checkPasswordStrength("always"))
print(checkPasswordStrength("always1234"))
print(checkPasswordStrength("Always1234"))
print(checkPasswordStrength("Always@123"))
print(checkPasswordStrength("Always/123"))
print(checkPasswordStrength("Always\\123"))
```
