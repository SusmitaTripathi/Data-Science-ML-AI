# **Loops**

1. For loop
2. While loop
3. Loop Control statements

##### **Range function**

```python
range(5) #generates a range of numbers from 0 - 4
```

```python
range(1,6) #generates range of numbers from 1-5
```

```python
range(1,10,2) #generates range of numbers from 1 - 9 with step size 2
```

#### **1. For Loop**

Iterating over a range of numbers

```python
for i in range(5):
    print(i)
```

```python
for i in range(1,10,2):
    print(i)
```

```python
for i in range(5,0,-1):
    print(i)
```

Iterating over a string

```python
string = "Susmita Tripathi"
for i in string:
    print(i)
```

#### **2. While Loop**

loop continues to execute while the condition is true

```python
count = 0
while count < 5:
    print(count)
    count += 1
```

```python
count = 0

while count%2 == 0:
    print(count)
    count += 1
```

#### **3. Loop Control Statements**

##### **i. Break Statement**

Break statement exists the loop prematurely

```python
for i in range(10):
    if i==5:
        break
    print(i)
```

##### **ii. Continue Statement**

skips the current iteration and continues with the next

```python
for i in range(10):
    if i%2 == 0:
        continue
    print(i)
```

##### **iii. Pass Statement**

pass statement is a null condition and does nothing

```python
for i in range(5):
    if i==3:
        print("The number is", i)
        pass
    print(i)
```

#### **Nested Loops**

Loop inside a loop

```python
for i in range(3):
    for j in range(2):
        print(f"i: {i}, j: {j}")
```

## **Examples**

##### **1. Calculate the sum of first n natural numbers using for and while loop**

For loop

```python
sum = 0
for i in range(11):
    sum += i
print("Sum of first 10 natural numbers:", sum)
```

While loop

```python
sum = 0
count = 1
while count<=10:
    sum += count
    count+=1
print("Sum of first 10 natural numbers:", sum)
```

##### **2. Prime numbers between 1 and 100**

```python
for num in range(1,101):
    if num > 1:
        for i in range(2,num):
            if num % i == 0:
                break
        else:
            print("The number", num,"is a prime number")
```
