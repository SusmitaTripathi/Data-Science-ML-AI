# **Conditional Statements**

#### **1. If statement**

Evaluates a condition and executes the block of code if condition is true



```python
age = 20

if age>=20:
    print("Age is greater than or equal to 20")
```

```python
age = 18

if age>=20:
    print("Age is greater than or equal to 20")
```

#### **2. If-Else statement**

Evaluates a condition and executes the if-block if condition is true or executes the else block



```python
age = 18

if age>= 20:
    print("Age is greater than or equal to 20")
else:
    print("Age is less than 20")
```

#### **3. Elif statement**

Allows to check multiple conditions == Else if



```python
age = int(input("Enter your age: "))
print("Age: ", age)
if age <= 13:
    print("Age is less than or equal to 13")
elif age <= 18:
    print("Age is less than or equal to 18")
elif age <= 20:
    print("Age is less than or equal to 20")
else:
    print("Age is greater than 20")
```

#### **4. Nested Conditional Statements**

Can place one or more if, elif, or else statements inside another  if, elif, or else statement.

```python
number = int(input("Enter a number: "))
print("Number: ", number)
if number > 0:
    if (number % 2) == 0:
        print("Number is even")
    else:
        print("Number is odd")
elif number == 0:
    print("Number is zero")
else:
    print("Number is negative")
```

## **Use Cases & Practical Examples**

#### **1. Check if number is leap year**

```python
year = int(input("Enter a year: "))
print("Year: ", year)
if year % 4 == 0:
    if year % 100 == 0:
        if year % 400 == 0:
            print("Year entered is a leap year")
        else:
            print("Year entered not leap year")
    else:
        print("Year entered is a leap year")
else:
    print("Year entered not leap year")
```
