# **Real-World Examples using Lists in Python**

**1. Manage a to-do list**
*     create a list to keep track of tasks

```python
toDoList = ["Buy groceries", "Clean the house", "Pay bills", "Go home"]
print(toDoList)
```

```python
#add task to toDoList
toDoList.append("Schedule meeting")
toDoList.append("Go for a run")
print(toDoList)
```

```python
#remove a task
toDoList.remove("Schedule meeting")
print(toDoList)
```

```python
#check if a task in the list
if 'Pay bills' in toDoList:
    print("Pay utility bills")
```

```python
#print tasks remaining
for task in toDoList:
    print(task)
```

```python
for task in toDoList:
    print(f"{task}")
```
