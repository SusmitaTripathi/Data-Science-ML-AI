## **List Comprehension**

##### **Basic syntax: [(expression) (for items in iterable)]**


```python
lst = []
for x in range(10):
    lst.append(x**2)
print(lst)
```

```python
square = [[x**2 for x in range(10)]] #using basic syntax
print(square)
```

##### **with conditional logic: [(expression) (for item in iterable) (if condition)]**

```python
lst=[]
for i in range(10):
    if i%2 == 0:
        lst.append(i)
print(lst)
```

```python
evenNumbers = [num for num in range(10) if num % 2 == 0] #using conditional logic syntax
print(evenNumbers)
```

##### **Nested List comprehension: [(expression) (for item1 in list 1) (for item 2 in list 2)]**

```python
lst1 = [1,2,3,4]
lst2 = ['a','b','c','d']

pair = [[i,j] for i in lst1 for j in lst2]
print(pair)
```

##### **List comprehension with function calls**

```python
words = ["hello","world","python","list","comprehension"]
lengths = [len(word) for word in words]
print(lengths)
```
