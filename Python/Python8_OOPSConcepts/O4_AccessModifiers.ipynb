# **Access Modifiers**

1. Public
2. Protected
3. Private

#### Public Variables

Variables can be accessed outside the class

```python
class Person:
    def __init__(self,name,age):
        self.name = name #declaring a public variable
        self.age = age #declaring a public variable

#creating object of class Person
person = Person("Name",21)

#accessing class variables directly
print(person.name)
print(person.age)
```

if you check directory of object - 'person' you will be able to see all publicly available methods and variables.
As name and person are public, they will also be in the list.

```python
dir(person)
```

#### **Private Variables**

Variables cannot be accessed outside the class. Not even in derived class.

```python
class PersonPrivate:
    def __init__(self,name,age,gender):
        self.__name = name #declaring a private variable
        self.__age = age #declaring a private variable
        self.gender = gender #public variable

#creating object of class Person
personPrivate = PersonPrivate("Name",21,'M')
```

```python
dir(personPrivate) #name,age not in the list as they are private
```

```python
print(personPrivate.name)
```

Above error as name is private and cannot be accessed from outside the class

Private variables can be accessed outside the class using below method but it is **_NOT RECOMMENDED_**.

```python
print(personPrivate._PersonPrivate__name)
```

#### **Protected Variables**

Cannot be accessed outside the class except for in derived classes.

```python
class PersonProtected:
    def __init__(self,name,age,gender):
        self._name = name #declaring a protected variable
        self._age = age #declaring a protected variable
        self.gender = gender #public variable

    @property
    def name(self):
        return self._name


#creating object of class Person
personProtected = PersonProtected("Name",21,'M')
```

```python
dir(personProtected)
```

```python
class Employee(PersonProtected):
    def __init__(self,name,age,gender):
        super().__init__(name,age,gender)

employee = Employee("Name",21,'M')
print(employee._name) #accessing a protected variable through a derived class object
```
