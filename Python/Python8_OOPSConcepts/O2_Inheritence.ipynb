#### **Inheritance In Python**

Inheritance is a fundamental concept in Object-Oriented Programming that allows a class to inherit attributes and methods from another class.

```python
#Parent class

class Car:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year

    def drive(self):
        print(f"Driving {self.make} {self.model} made in the year {self.year}")
```

```python
car1 = Car("BMW","X Class","2022")
car1.drive()
```

#### **Single Inheritance**

Child class inherits the parent class

**`class childClassName(ParentClassName):`**


```python
#child class

class Tesla(Car):
    def __init__(self, make, model, year,isSelfDriving = True):
        super().__init__(make, model, year)
        self.isSelfDriving = isSelfDriving

    def selfDriving(self):
        if self.isSelfDriving:
            print(f"Tesla {self.make}  {self.model} supports self driving")
        else:
            print(f"Tesla {self.make} {self.model} does not support self driving")
```

```python
tesla1 = Tesla("XG","Electric Engine",2024,True)
tesla1.selfDriving()
```

#### **Multiple Inheritance**

When a class inherits from more than one class.

```python
#Base class 1
class Animal:
    def __init__(self, name):
        self.name = name

    def describe(self):
        print(f"Animal is {self.name}")
```

```python
#Base class 2
class Pet:
    def __init__(self, parent):
        self.parent = parent

    def petDescription(self):
        print(f"Pet Parent is {self.parent}")

    def getDetails(self):
        print("Print from class Pet")
```

```python
class Dog(Animal, Pet):
    def __init__(self, name,parent):
        Animal.__init__(self,name)
        Pet.__init__(self,parent)

    def getDetails(self):
        Animal.describe(self)
        Pet.petDescription(self)
```

```python
dog1 = Dog("Husky","Maya")
dog1.getDetails()
```
