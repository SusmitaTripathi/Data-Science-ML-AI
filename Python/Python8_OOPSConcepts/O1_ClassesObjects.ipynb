# **Classes and Objects**

Object-Oriented Programming (Oop) is a programming paradigm that uses "objects" to design applications and computer programs. OOP allows for modeling real-world scenarios using classes and objects.

#### **Classes**

is a blueprint for creating objects. Has its own attributes and methods

Based on the class, we can create different objects that use the features of the class.

```python
#empty class
class Car:
    pass

#objects of class Car
audi = Car()
bmw = Car()

print(type(audi))
print(type(bmw))
```

#### **Constructor**

Can define all the attributes of the class can be defined using constructor.

Syntax:

`def __init__(parameters):`

```python
class Dog:
    def __init__(self, name, age):
        self.name = name # self.name = variable of the class # name = value passed while creating the objects
        self.age = age

dog1 = Dog("Buddy",3)
print(dog1.name)
print(dog1.age)

dog2 = Dog("Honey",2)
print(dog2.name)
print(dog2.age)
```

Class definition with instance methods

```python
class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def bark(self):
        print(f"{self.name} barks and is {self.age} years old")

dog3 = Dog("Buddy",3)
print(dog3.name)
print(dog3.age)
dog3.bark()

dog4 = Dog("Honey",2)
print(dog4.name)
print(dog4.age)
dog4.bark()
```

#### **Modelling a bank account**

```python
class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.balance = balance

    def deposit(self,amount):
        self.balance += amount
        print(f"{self.owner} deposited {amount} \nTotal Balance: {self.balance}")

    def withdraw(self,amount):
        if amount < self.balance:
            self.balance -= amount
            print(f"{self.owner} withdrawn {amount} \nTotal Balance: {self.balance}")
        else:
            print("Insufficient Balance")

    def getBalance(self):
        print(f"Total Balance: {self.balance}")
```

```python
account1 = BankAccount("Susmita",5000)
```

```python
account1.deposit(100)
account1.getBalance()
```

```python
account1.withdraw(200)
account1.getBalance()
```
